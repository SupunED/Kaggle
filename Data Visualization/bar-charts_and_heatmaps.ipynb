```python
import pandas as pd
pd.plotting.register_matplotlib_converters
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print("Setup Complete")
```

```python
# path of the dataset
flight_filepath = "flight_delays.csv"
# read the dataset to a variable
flight_data = pd.read_csv(flight_filepath)
```

```python
flight_data
```

```python
# set the width and height of the figure
plt.figure(figsize=(10,6))

# add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.Month, y=flight_data['NK'])

# add label for vertical axis
plt.ylabel("Arrival delay in minutes")
# add label for horizontal axis
plt.xlabel("Month")
```

```python
# set the width and the height for the fig
plt.figure(figsize=(14,7))

# add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# add label for the horizontal axis
plt.xlabel("Airliner")
```
