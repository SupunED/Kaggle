```python
# importing modules and setting up the env
import pandas as pd
pd.plotting.register_matplotlib_converters
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# specify data location and load the dataset
spotify_filepath = "spotify.csv"
spotify_data = pd.read_csv(spotify_filepath)
```

```python
# read the first 5 rows of data
spotify_data.head()
```

```python
# read the last 5 rows of data
spotify_data.tail()
```

```python
# Line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)
```

```python
# set the width and the height of the figure
plt.figure(figsize=(14,6))

# add title
plt.title("Daily global streams of popular songs in 2017-2018")

# Line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)
```
