```python
import pandas as pd
```

```python
data = pd.read_csv('SAP-4000.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe()
```

```python
df['Gender'] = pd.to_numeric(df['Gender'].replace({'Male': 1, 'Female': 0}, regex=True), errors='coerce')
df['Tutoring'] = pd.to_numeric(df['Tutoring'].replace({'Yes': 1, 'No': 0},regex=True), errors='coerce')
df['Region'] = pd.to_numeric(df['Region'].replace({'Rural': 1, 'Urban': 0},regex=True), errors='coerce')
df['Parent Education'] = pd.to_numeric(df['Parent Education'].replace({'None' : 0, 'Primary' : 1, 'Secondary' : 2, 'Tertiary' : 3},regex=True), errors='coerce')
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['Exam_Score']
features = ['HoursStudied/Week', 'Attendance(%)', 'Region', 'Gender', 'Tutoring', 'Parent Education']
X = df[features]
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
SAP_model = DecisionTreeRegressor()
```

```python
SAP_model.fit(train_X, train_y)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(val_y)
```

```python
predictions = SAP_model.predict(val_X)
mae = mean_absolute_error(val_y, predictions)
print(predictions)
```

```python
print(f'mean absolute error = {mae}')
```
