```python
import pandas as pd
```

```python
vehicle_data = pd.read_csv('vehicle.csv')
```

```python
df = pd.DataFrame(vehicle_data)
```

```python
df.head(3)
```

```python
df['Price'] = pd.to_numeric(df['Price'].replace({'Rs ': '', 'Rs. ': '', ',': '', ' ': ''}, regex=True), errors='coerce')
df['Price']
```

```python
df['Capacity'] = pd.to_numeric(df['Capacity'].replace({'cc': '', ',': '', ' ': ''}, regex=True), errors='coerce')
df['Capacity']
```

```python
df['Mileage'] = pd.to_numeric(df['Mileage'].replace({' ': '', ' km': '', ',': ''}, regex=True), errors='coerce')
df["Mileage"]
```

```python
df['Year'] = pd.to_numeric(df["Year"])
df['Year']
```

```python
y = df['Price']
features = ['Year','Capacity','Mileage']
X = df[features]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
vehicle_model = RandomForestRegressor(random_state=1)
```

```python
vehicle_model.fit(train_X, train_y)
```

```python
price_prediction = vehicle_model.predict(val_X)
print(val_y, price_prediction)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(val_y, price_prediction)
```
